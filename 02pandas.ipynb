```python
managers = pd.read_csv('data/managers.csv')
```

## 국적이 '미국'인 매니저의 나이는?

```python
where = managers.Country == 'US'
```

```python
# 조건식의 결과가 참인 행만 성택해서 출력
managers[where]
```

```python
managers[where][['Age']]
```

### 성별이 '남자'인 매니저의 국적과 나이는?

```python
where = managers.Gender == 'M'
managers[where][['Country','Age']]
managers[where].loc[:,['Country','Age']]
managers[where].iloc[:,[2,4]]
```

### 핸드폰 구매자 데이터

```python
phone = pd.read_csv('data/phone02.csv')
phone.head()
```

```python
# 구매자의 평균 나이
amean = phone.mean()['age']
amean
```

```python
# 구매자의 평균 몸무게
wmean = phone.mean()['weight']
wmean
```

```python
# 구매자의 평균 키 
hmean = phone.mean()['height']
hmean
```

```python
# 연도별 추출 - 출시연도가 2015인 정보 추출
where = phone['makeyear'] == 2015
phone[where]
```

```python
# 연도별 추출 - 구매연도가 2016인 정보 추출
where = phone['buyyear'] == 2016
phone[where]
```

```python
# 나이별 추출 - 나이가 20 ~ 30사이인 구매자 추출
#where = (phone['age'] >= 20) & (phone['age'] <= 30)
where = phone.age.between(20,30)
phone[where]
```

```python
# 화면크기가 4.5인 핸드폰을 구매한 소비자의 나이, 키, 몸무게 조회
where = phone['dispsize'] == 4.5
phone[where].loc[:,['age','height','weight']]
```

```python
# 나이가 35세이상인 소비자들이 구매한 핸드폰의 출시연도, 화면크기 조회
where = phone['age'] >= 35
phone[where].loc[:,['makeyear','dispsize']]
```
